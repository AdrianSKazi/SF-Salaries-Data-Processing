#Amount of people who have Chief word in their job title

```python
def chief(x):
  if 'chief' in x.lower().split():
    return True
  else:
    return False
```

```python
pd.DataFrame(sale['JobTitle'].apply(lambda x: chief(x)))['JobTitle'].value_counts().iloc[1]
```

#Correlation between length of Job Title string and Total Pay

```python
sale['title_len'] = sale['JobTitle'].apply(len)
corr = sale[['TotalPayBenefits','title_len']].corr(method='pearson')
corr
```

```python
coef_of_determinatin = corr**2
coef_of_determinatin
```

Correlation indicator and coefficient of determination are so neglible that we can't tell that there is any correlation between Job title length and Total Pay.
